# Q2.  Scrape first 3 pages from amazon.in for Powerbanks.

```python
#Lets first connect to web driver
driver=webdriver.Chrome("C:\web driver\chromedriver.exe")
```

```python
driver.get('https://www.amazon.in/')
```

```python
#finding element for power banks in search bar

driver.find_element_by_xpath("//div[@class='nav-search-field ']/input").send_keys(input)
```

```python
#pressing the the search button.
driver.find_element_by_xpath("//div[@class='nav-search-submit nav-sprite']").click()
```

```python
#specifing the url of the webpage to be scraped
url="https://www.amazon.in/s?k=power+banks&ref=nb_sb_noss_2"
```

```python
urla=driver.find_elements_by_xpath("//ul[@class='a-pagination']//li/a")
urls=[i.get_attribute('href') for i in urla[0:3]]
urls
```

```python
link=[]
for i in urls:[]
    driver.get(i)
    for j in driver.find_elements_by_xpath("//a[@class='a-link-normal s-no-outline']"):
        link.append(j.get_attribute('href'))
link
```

```python
np=[]
nr=[]
pc=[]
rat=[]
de=[]
av=[]
br=[]
re=[]
od=[]
url=[]
for i in link:
    driver.get(i)
    
#Name of Product
    try:
        np.append(driver.find_element_by_xpath("//span[@id='productTitle']").text)
    except NoSuchElementException as e:
        np.append("--")
        
#No of ratings
    try: 
        nr.append(driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']").text)
    except NoSuchElementException as e:
        nr.append("--")
        
#Price
    try:
        pc.append(driver.find_element_by_xpath("//td[@class='a-span12']/span").text)
    except NoSuchElementException as e:
        pc.append("--")
        
#Locating Rating
    
    try:
        rat.append(driver.find_element_by_xpath("//span[@data-hook='rating-out-of-text']").text)
    except NoSuchElementException as e:
        rat.append("--")
#delivery

    try:
        de.append(driver.find_element_by_xpath("//div[@id='ddmDeliveryMessage']/b").text)
    except NoSuchElementException as e:
        de.append("--")
        
#availability
    try:
        av.append(driver.find_element_by_xpath("//div[@id='availability']/span").text)
    except NoSuchElementException as e:
        av.append("--")
        
#brand
    try:
        br.append(driver.find_element_by_xpath("//div[@class='a-section a-spacing-none']/a").text)
    except NoSuchElementException as e:
        br.append('--')
        
#return/exchange
    try:
        re.append(driver.find_element_by_xpath("//div[@id='RETURNS_POLICY']").text)
    except NoSuchElementException as e:
        re.append("--")

#urls

    url.append(i)
    
#other details

    try:
        od.append(driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']").text)
    except NoSuchElementException as e:
        od.append("--")



```

```python
len(link)
```

```python
# Creating a dataframe for powerbanks. 

powerbanks=pd.DataFrame({})
powerbanks['Brand']=br
powerbanks['Name Of Product']=np
powerbanks['Rating']=rat
powerbanks['No of Ratings']=nr
powerbanks['Price']=pc
powerbanks['Return/Exchange']=re
powerbanks['Expected Delivery']=de
powerbanks['Availability']=av
powerbanks['Other Details']=od
powerbanks['Product URL']=url

powerbanks
```

```python
driver.close()
```

# Q4.  Flipkart scrape all the details of a smartphone.

```python
#Lets first connect to web driver
driver=webdriver.Chrome("C:\web driver\chromedriver.exe")
```

```python
driver.get('https://www.flipkart.com/')
```

```python
#finding element for pixel 4a in search bar

driver.find_element_by_xpath("//input[@class='_3704LK']").send_keys('pixel 4A')
driver.find_element_by_xpath("//button[@class='L0Z3Pu']").click()
```

```python
url="https://www.flipkart.com/search?q=pixel%204A&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"
```

```python
url=[]
for i in driver.find_elements_by_class_name('_1fQZEK'):
    url.append(i.get_attribute('href'))
url
```

```python
md=[]
col=[]
pr=[]
ram=[]
rom=[]
pri=[]
sec=[]
dis_siz=[]
dis_res=[]
pro=[]
core=[]
bat=[]
urls=[]

for i in url:
    driver.get(i)
    
# Model
    driver.find_element_by_xpath("//button[@class='_2KpZ6l _1FH0tX']").click()
    time.sleep(2)
    try:
        md.append(driver.find_element_by_xpath("//table[@class='_14cfVK']//tr[3]//td//ul").text)
    except:
        md.append('--')
        
# colour

    try:
        col.append(driver.find_element_by_xpath("//table[@class='_14cfVK']//tr[4]//td//ul").text)
    except NoSuchElementException as e:
        col.append('--')

#price
    
    try:
        pr.append(driver.find_element_by_xpath("//div[@class='_30jeq3 _16Jk6d']").text)
    except NoSuchElementException as e:
        pr.append('--')
        
#ram

    try:
        ram.append(driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[4]//tr[2]//ul//li").text)
    except NoSuchElementException as e:
        ram.append('--')
        
#rom

    try:
        rom.append(driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[4]//ul/li").text)
    except NoSuchElementException as e:
        rom.append('--')
        
#primary camera 
     
    try:
        pri.append(driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[5]//tr[2]//ul//li").text)
    except NoSuchElementException as e:
        pri.append('--')
        
#secondary camera

    try:
        sec.append(driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[5]//tr[5]//ul//li").text)
    except NoSuchElementException as e:
        sec.append('--')
        
#Display size

    try:
        dis_siz.append(driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[2]//tr[1]//ul//li").text)
    except NoSuchElementException as e:
        dis_siz.append('--')

#Display resolution

    try:
        dis_res.append(driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[2]//tr[1]//ul//li").text)
    except NoSuchElementException as e:
        dis_res.append('--')
        
#Processor 

    try:
        pro.append(driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[3]//tr[1]//ul//li").text)
    except NoSuchElementException as e:
        pro.append('--')

#Processor Core

    try:
        core.append(driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[3]//tr[3]//ul//li").text)
    except NoSuchElementException as e:
        core.append('--')
        
#Battery

    try:
        bat.append(driver.find_element_by_xpath("//div[@class='_1UhVsV']//div[9]//ul//li").text)
    except NoSuchElementException as e:
        bat.append('--')
        
#URL

    urls.append(i)        
```

```python
br=[]
for i in url:
    driver.get(i)
    
#Brand name
    try:
        driver.find_element_by_class_name('_3IIjhd').click()
        time.sleep(3)
        br.append(driver.find_element_by_xpath("//section[@class='_1-ermZ']/h4").text)
        driver.find_element_by_xpath("//div[@class='_2_e-g9 _28WHFR']/button").click()
        time.sleep(5)
    except NoSuchElementException as e:
        br.append("--")
```

```python
 br
```

```python
# Creating a dataframe for  google pixel4a. 
pixel4a=pd.DataFrame({})
pixel4a['Brand']=br
pixel4a['Model']=md
pixel4a['Colour']=col
pixel4a['Price']=pr
pixel4a['Ram']=ram
pixel4a['Rom']=rom
pixel4a['Primary Camera']=pri
pixel4a['Secondary Camera']=sec
pixel4a['Display Size']=dis_siz
pixel4a['Display Resolution']=dis_res
pixel4a['Processor']=pro
pixel4a['Processor Core']=core
pixel4a['Battery']=bat
pixel4a['Product Url']=urls

pixel4a
```

```python
# Creating a dataframe for trak.in
trak=pd.DataFrame({})
trak['Date']=date
trak['Startup Name']=start
trak['Industry/Vertical']=vertical
trak['Sub-Vertical']=sub
trak['City/Location']=city
trak['Investors Name']=invest
trak['Investment Type']=investtype
trak['Amount(In USD)']=amt

trak
```

# Q7. Write a program to scrap all the available details of top 10 gaming laptops from digit.in. 

```python
#Lets first connect to web driver
driver=webdriver.Chrome("C:\web driver\chromedriver.exe")
```

```python
driver.get("https://www.digit.in/")
```

```python
# Next Clicking on the element for Top 10 Gaming Laptops.

driver.find_element_by_xpath("//div[@class='listing_container']//li[9]/a").click()
```

```python
# Specifing the url of the webpage to be scraped
url="https://www.digit.in/top-products/best-gaming-laptops-40.html"
```

```python
# Creating empty lists for the information of the laptops to be scraped form above url.

nam=[]
pri=[]
slr=[]
os=[]
dis=[]
pro=[]
mry=[]
wgt=[]
dim=[]
grp=[]


#Product Name

for z in driver.find_elements_by_tag_name('h3'):
    nam.append(z.text)
    
#Price

for y in driver.find_elements_by_xpath("//td[@class='smprice']"):
    pri.append(y.text)
    
#Seller

for x in driver.find_elements_by_xpath("//td[@class='smmerchant']"):
    slr.append(x.text)

#Operating System

for w in driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[3]//td[3]"):
    os.append(w.text)
    
#Display

for v in driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[4]//td[3]"):
    dis.append(v.text)
    
#Processor

for u in driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[5]//td[3]"):
    pro.append(u.text)
    
#Memory

for t in driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[6]//td[3]"):
    mry.append(t.text)
    
#Weight

for s in driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[7]//td[3]"):
    wgt.append(s.text)
    
#Dimension

for r in driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[8]//td[3]"):
    dim.append(r.text)
    
#Graphics Processor

for q in driver.find_elements_by_xpath("//div[@class='Spcs-details']//tr[9]//td[3]"):
    grp.append(q.text)
```

```python
len(grp)
```

```python
# Extracting hrefs of laptops to scrape out more information
url=[]
for i in driver.find_elements_by_xpath("//div[@class='TopNumbeHeading sticky-footer' or @class='TopNumbeHeading active sticky-footer' or @class='TopNumbeHeading sticky-footer active']//a"):
    url.append(i.get_attribute('href'))
url
```

```python
web=[]
stat=[]
date=[]
pros=[]
cons=[]
spec=[]

for i in url:
    driver.get(i)
    
#Market Status

    try:
        stat.append(driver.find_element_by_xpath("//div[@class='Block-status']/b").text)
    except NoSuchElementException as e:
        stat.append("--")
        
#Website

    try:
        web.append(driver.find_element_by_xpath("//div[@class='status-bar only-desktop']//div[3]/b").text)
    except NoSuchElementException as e:
        web.append("--")

#Launch Date

    try:
        date.append(driver.find_element_by_xpath("//div[@class='status-bar only-desktop']//div[2]/b").text)
    except NoSuchElementException as e:
        date.append("--")
        
#Pros

    try:
        pros.append(driver.find_element_by_xpath("//div[@class='pros-Cons']//ul[1]").text.strip())
    except NoSuchElementException as e:
        pros.append("--")
        
#Cons

    try:
        cons.append(driver.find_element_by_xpath("//div[@class='pros-Cons']//ul[2]").text.strip())
    except NoSuchElementException as e:
        cons.append("--")
        
#Key Specifications

    try:
        spec.append(driver.find_element_by_xpath("//div[@class='Key-specs']").text.strip())
    except NoSuchElementException as e:
        spec.append("--")
```

```python
len(cons)
```

```python
# Creating Dataframe for Top 10 Gaming Laptops from Digit.in

GamingLaptop=pd.DataFrame({})
GamingLaptop['Name']=nam
GamingLaptop['Price']=pri
GamingLaptop['Seller']=slr
GamingLaptop['Operating System']=os
GamingLaptop['Display']=dis
GamingLaptop['Processor']=pro
GamingLaptop['Memory']=mry
GamingLaptop['Weight']=wgt
GamingLaptop['Dimension']=dim
GamingLaptop['Graphics Processor']=grp
GamingLaptop['Website']=web
GamingLaptop['Market Status']=stat
GamingLaptop['Release Date']=date
GamingLaptop['Pros']=pros
GamingLaptop['Cons']=cons
GamingLaptop['Key Specifications']=spec

GamingLaptop
```

```python
driver.close()
```
